```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('./Dataset/Dataset.csv', encoding='ISO-8859-1')
```

```python
dataframe
```

```python
keywords_expanded = dataframe['Keywords'].str.split('|', expand=True)
keywords_expanded
```

```python
dataframe_keywords = pd.get_dummies(keywords_expanded.stack())
dataframe_keywords
```

```python
dataframe = pd.concat([dataframe, dataframe_keywords], axis = 1)
dataframe
```
